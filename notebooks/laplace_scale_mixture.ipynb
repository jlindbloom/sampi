```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def neg_log_joint_dens_func(z, s, a=1.0):
    """Evaluates the log of the joint density function for (z,s).
    """
    const = -np.log(1/np.sqrt(s))
    t1 = (1.0/(2*s))*((z**2)/(2*s))
    t2 = s*((a**2)/2)
    return const + t1 + t2
```

```python

# a parameter
a = 1.0

# Generate data points for x and y
z = np.linspace(-5, 5, 100)
s = np.linspace(1e-1, 3, 100)

# Create a meshgrid from x and y
Z, S = np.meshgrid(z, s)

# Evaluate the function for each combination of x and y
W = np.exp(-neg_log_joint_dens_func(Z, S, a=a))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(Z, S, W, cmap='viridis')

# Add colorbar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels for axes
ax.set_xlabel('Z-axis')
ax.set_ylabel('S-axis')
ax.set_zlabel('W-axis')

# Set the title for the plot
ax.set_title('Joint density for (z,s)')

ax.view_init(elev=30, azim=10)

ax.grid(False)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define your custom function
def func(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Generate data points for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a meshgrid from x and y
X, Y = np.meshgrid(x, y)

# Evaluate the function for each combination of x and y
Z = func(X, Y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add colorbar
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels for axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set the title for the plot
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```
